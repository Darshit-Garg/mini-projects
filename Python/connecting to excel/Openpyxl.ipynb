```python
import openpyxl 
from openpyxl.styles import Font

def add():
    n = input('Enter number of students: ')
    lst = []
    wb = openpyxl.load_workbook('student.xlsx')
    sheet = wb.active
    for _ in range(int(n)):  
        lst = []
        row = sheet.max_row
        for i in range(1, row + 1):
            cell = sheet.cell(row=i, column=1)
            lst.append(cell.value)
#        print(lst)
        while True:
            ID = input('Enter 4 digit ID: ')
            if ((len(ID)!=4) or (ID.isnumeric == False) or (ID in lst)):
                print('Please enter new valid ID')
            else:
                break
        while True:            
                list = [ID]
                name = input('Enter name: ')
                list.append(name)
                try:
                    for i in range(1,5):
                        M = float(input(f'Enter Marks {i}: '))
                        if ((M<0) or (M>100)):
                            print('Please Enter Valid Marks')
                            i -= 1
                        else:    
                            list.append(M)
                    break
                except ValueError:
                    print('Enter Valid data')
        tup = tuple(list)
#        print(tup)
        flist = []
        flist.append(tup)
#        print(flist)
        ftup = tuple(flist)
        for row in ftup:    
            sheet.append(row)
        wb.save('student.xlsx')


def remove():
    wb = openpyxl.load_workbook('student.xlsx')
    sheet = wb.active
    rows = sheet.max_row
    lst = []
    for i in range(1,rows+1):
        cell = sheet.cell(row = i, column = 1)
        lst.append(cell.value) 
    while True:
        rtbr = input('Enter ID: ')
        if rtbr in lst:
            break
        else:
            print('Please Enter an already existing ID')
    index = lst.index(rtbr)
    index+=1
    sheet.delete_rows(index)
    wb.save('student.xlsx')
        

def update():
    wb = openpyxl.load_workbook('student.xlsx')
    sheet = wb.active
    lst = []
    row = sheet.max_row
    for i in range(1, row + 1):
        cell = sheet.cell(row=i, column=1)
        lst.append(cell.value)
    while True:    
        ID = input('Enter the ID whose data has to be updated: ')
        if ID not in lst:
            print('Enter valid ID')
        else:
            try:
                row = lst.index(ID)
                print('Enter the data: ')
                name = input('Enter name: ')
                cel = sheet.cell(row = row+1, column = 2)
                cel.value = name
                M1 = float(input('Enter Marks1: '))
                cel = sheet.cell(row = row+1, column = 3)
                cel.value = M1
                M2 = float(input('Enter Marks2: '))
                cel = sheet.cell(row = row+1, column = 4)
                cel.value = M2
                M3 = float(input('Enter Marks3: '))
                cel = sheet.cell(row = row+1, column = 5)
                cel.value = M3
                M4 = float(input('Enter Marks4: '))
                cel = sheet.cell(row = row+1, column = 6)
                cel.value = M4
                break
            except ValueError:
                print('Enter Valid Data')
    wb.save('student.xlsx')
        
    
def search():
    name = []
    ID = []
    wb = openpyxl.load_workbook('student.xlsx')
    sheet = wb.active
    rows = sheet.max_row
    while True:    
        choice = input('Search by name (0)?\nSearch by ID (1)?')
        if ((choice!=1) or (choice!=0)):
            print('Please enter among 1 or 0')
        else: 
            break
    for i in range(1, rows + 1):
        cell = sheet.cell(row=i, column=2)
        name.append(cell.value)
    for i in range(1, rows + 1):
        cell = sheet.cell(row=i, column=1)
        ID.append(cell.value)
    while True: 
        if choice == 0:    
            Name = input('Enter name: ')
            if Name not in name:
                print('Enter valid name')
            else: 
                row = name.index(Name)
                row+=1
                for i in range(1,7):
                    cell = sheet.cell(row = row, column = i)
                    print(cell.value,end = ' ')
                print()
                break
        if choice == 1:    
            Id = input('Enter ID: ')
            if Id not in ID:
                print('Enter valid ID')
            else: 
                row = ID.index(Id)
                row+=1
                for i in range(1,7):
                    cell = sheet.cell(row = row, column = i)
                    print(cell.value,end = ' ')
                print()
                break
    wb.save('student.xlsx')


if __name__ == '__main__':
    print('Student Data'.center(120))
    print('Welcome!')
    wb = openpyxl.load_workbook('student.xlsx')
    sheet = wb.active
    c = sheet['A2']
    check = 1
    if (c.value == None):
        print('What would you like to do?\n')
        print('Add data (1)\nExit (5)')
        while True:    
            str = input()
            if str.strip() == '1':
                add()
                break
            elif str.strip() == '5':
                check = 0
                break 
            else:
                print("Enter among '1','5'")
#    n = input('Enter number of students: ')    
#    add(n)    
    if check == 1:
        print('What would you like to do?\n')
        print('Add data (1)\nRemove data (2)\nUpdate data (3)\nSearch (4)\nExit (5)')
        while True:    
            str = input()
            if str.strip() == '1':
                add()
                break
            elif str.strip() == '2':
                remove()
                break
            elif str.strip() == '3':
                update()
                break
            elif str.strip() == '4':
                search()
                break
            elif str.strip() == '5':
                break
            else:
                print("Enter among '1','2','3','4','5'")
    print('Thank you')
```
